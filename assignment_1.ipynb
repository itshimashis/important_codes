```python
rows=149
cols=149
sim_matrix=[]
```

```python
import math

# Assuming 'arr' is a 2D array containing the data and 'matrix' is a 2D array for storing the distances
# Make sure to initialize the 'matrix' array before running this code

for i in range(149):
    temp=[]
    for j in range(149):
        val = 0
        for k in range(4):
            val += (data1[i][k] - data1[j][k]) * (data1[i][k] - data1[j][k])
        val = math.sqrt(val)
        temp.append(val)

    sim_matrix.append(temp)
```

```python
df=pd.DataFrame(sim_matrix)
df
```

```python
#now calculating the average similarity for every element
#in our case the lesser is better.
average=[]
for i in range(149):
    val=0
    for j in range(149):
        val+=sim_matrix[i][j]

    val=val/149
    average.append(val)
```

```python
average
```

```python
clusters=[]
for i in range(149):
    temp=[]
    for j in range(149):
        if(average[i]<=sim_matrix[i][j]):
            temp.append(j)

    clusters.append(temp)
```

```python
#now we have to remove some clusters
#lets say there are clusters A and B and if A contains all the elements of
#B and len(A)>=len(B) that means B is subset cluster of A
def f(a, b):
    a.sort()
    b.sort()

    if len(a) < len(b):
        return False #is a is lesser in size then b cannot be a subcluster to it

    for i in b:
        if i not in a:
            return False

    return True
```

```python
#clusters.sort(key=lambda x: len(x), reverse=True) #taking the maxsize clusters in the front
duplicates=[]
for i in range(149):
    duplicates.append(False)
for i in range(149):
    if(duplicates[i]):
        continue;
    for j in range(149):
        if(i==j):
            continue
        if(f(clusters[i],clusters[j])):
            duplicates[j]=True
            print(j,"is a subcluster of",i)
```

```python
def intersection(a, b):
    ans = 0
    for i in a:
        if i in b:
            ans += 1
    return ans

def unionval(a, b):
    intersect = intersection(a, b)
    return len(a) + len(b) - intersect
```

```python
#now for each of this cluster we need to find the similarity value with every other cluster
#creating a 15*15 similarity matrix
C=[] #similarity matrix for the clusters
for i in range(15):
    temp=[]
    for j in range(15):
        if(i==j):
            temp.append(1)
        else:
            temp.append(intersection(unique_clusters[i],unique_clusters[j])/unionval(unique_clusters[i],unique_clusters[j]))

    C.append(temp)


```

```python
df5=pd.DataFrame(C)
df5
```

```python
def merge(a, b):
    for i in b:
        if i not in a:
            a.append(i)
    return a
```
